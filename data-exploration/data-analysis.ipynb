```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('property_data.csv')
```

```python
# Drop irrelevant columns
df = df.drop(['Raw num:', 'Type of Sale', 'URL'], axis=1)

# Convert columns to appropriate data types if needed
# ...

# Handle missing values if needed
# ...

# Print a 3-line preview of the cleaned DataFrame
display(df.head())
```

## 3. Visualization experiments

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot of price against zip code
sns.scatterplot(data=df, x='Zip code', y='Price of property in euro')
plt.title('Price vs. Zip Code')
plt.show()
```

```python
# Create a scatter plot of price against living area
sns.scatterplot(data=df, x='Living area', y='Price of property in euro')
plt.title('Price vs. Living Area')
plt.show()
```
